```python
import matplotlib.pyplot as plt
from math import log
import numpy as np



#The observed he actual number of transistors on CPUs between the years 1972 and 2012.
yr = [1972, 1974 , 1978 ,1982, 1985, 1989 ,1993, 1997, 1999 ,2000 , 2003 ,2004 ,2007 , 2008 , 2012]
tr = [0.0025 , 0.005, 0.029 , 0.12 , 0.275 , 1.18 , 3.1 , 7.5, 24.0 , 42.0, 220.0, 592.0 , 1720.0, 2046.0, 3100.0]


#Computes the Moore Law calculation.
def law(year):
    x = 0.0025 * 2**((year-1972)/2) 
    
    return x



lst = [] #All the predicted values are appended to this list.

# List iterate through each year and apply the moore law computation, and appends to the list.
for x in range(len(yr)):
    lst.append(law(yr[x])) 
lst = np.array(lst)


#Plot both observed and predicted values on graph
plt.scatter(yr,np.log(tr), marker = (5,2)) 
plt.plot(yr,np.log(lst), '--or') 
plt.title('Moore Law ') 
plt.legend(['Observed','Predicted']) 
plt.xlabel('Year') 
plt.ylabel('log(n_trans)') 

plt.show()

```
